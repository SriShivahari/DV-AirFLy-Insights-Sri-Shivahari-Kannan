```python
# %%
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load preprocessed dataset
df = pd.read_csv("DV_AirFly_Preprocessed.csv")

print("✅ Preprocessed dataset loaded. Shape:", df.shape)
df.head()
```

```python
# %%
# Top 10 airlines by flight count
top_airlines = df['AIRLINE'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_airlines.index, y=top_airlines.values, palette="Blues_r")
plt.title("Top 10 Airlines by Flight Count")
plt.xlabel("Airline")
plt.ylabel("Number of Flights")
plt.show()

# Top 10 routes by flight count
top_routes = df['Route'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_routes.index, y=top_routes.values, palette="Greens_r")
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Routes by Flight Count")
plt.xlabel("Route")
plt.ylabel("Number of Flights")
plt.show()

# Flights by month
monthly_flights = df['Month'].value_counts().sort_index()

plt.figure(figsize=(10,6))
sns.lineplot(x=monthly_flights.index, y=monthly_flights.values, marker="o")
plt.title("Flights by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()
```

```python
# %%
day_names = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
flights_per_day = df['DayOfWeek'].value_counts().sort_index()

plt.figure(figsize=(10,6))
sns.barplot(x=day_names, y=flights_per_day.values, palette="coolwarm")
plt.title("Flight Distribution by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Flights")
plt.show()
```

```python
# %%
flights_per_hour = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.barplot(x=flights_per_hour.index, y=flights_per_hour.values, palette="magma")
plt.title("Flight Distribution by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Flights")
plt.show()
```

```python
# %%
top_airports = df['ORIGIN'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_airports.index, y=top_airports.values, palette="viridis")
plt.title("Top 10 Origin Airports by Flight Count")
plt.xlabel("Airport Code")
plt.ylabel("Number of Flights")
plt.show()
```

```python
# %%
plt.figure(figsize=(12,5))
sns.histplot(df['ARR_DELAY'], bins=100, kde=False, color="steelblue")
plt.xlim(-50, 300)  # zoom in for readability
plt.title("Arrival Delay Distribution")
plt.xlabel("Arrival Delay (minutes)")
plt.ylabel("Flight Count")
plt.show()

plt.figure(figsize=(12,5))
sns.boxplot(x=df['ARR_DELAY'], color="salmon")
plt.xlim(-50, 300)
plt.title("Arrival Delay Boxplot")
plt.xlabel("Arrival Delay (minutes)")
plt.show()
```

```python
# %%
delay_causes = ["DELAY_DUE_CARRIER", "DELAY_DUE_WEATHER", "DELAY_DUE_NAS",
                "DELAY_DUE_SECURITY", "DELAY_DUE_LATE_AIRCRAFT"]

delay_by_airline = df.groupby("AIRLINE")[delay_causes].sum()

delay_by_airline.plot(kind="bar", stacked=True, figsize=(14,7), colormap="tab20c")
plt.title("Delay Causes by Airline")
plt.ylabel("Total Delay Minutes")
plt.xlabel("Airline")
plt.legend(title="Delay Cause", bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
# %%
avg_delay_by_hour = df.groupby("Hour")["ARR_DELAY"].mean()

plt.figure(figsize=(12,6))
sns.lineplot(x=avg_delay_by_hour.index, y=avg_delay_by_hour.values, marker="o", color="darkred")
plt.title("Average Arrival Delay by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Average Arrival Delay (minutes)")
plt.show()
```

```python
# %%
avg_delay_airports = df.groupby("ORIGIN")["ARR_DELAY"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_delay_airports.index, y=avg_delay_airports.values, palette="rocket")
plt.title("Top 10 Origin Airports by Average Arrival Delay")
plt.xlabel("Airport")
plt.ylabel("Average Arrival Delay (minutes)")
plt.show()
```
